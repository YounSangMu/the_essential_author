```python
import requests
from bs4 import BeautifulSoup
 
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://event.kyobobook.co.kr/desntl/detail/16',headers=headers)
 
soup = BeautifulSoup(data.text, 'html.parser')
title = soup.select('div.cont_txt')

for idx, div in enumerate(title, start=1):
    print(f"문단 {idx}: {div.get_text().strip()}")
```
